```python
 data_dict={
     "sales":[10,20,30,40],
     "profits":[2,4,3,-1],
     "date":["2022-01-01","2022-09-09","2022-08-01","2022-09-21"]
 }

df=pd.DataFrame(
    data_dict,
    columns=["sales","profits"],
             index=data_dict["date"]
             )
df
```

```python

fig,ax=plt.subplots()
ax.plot(df)
ax.set_title("product sales and profits")
```

```python

fig,ax=plt.subplots()
ax.plot(df.index,df["sales"])
ax.plot(df.index,df["profits"],ls="--")
ax.set_title("product sales and profits")
```
