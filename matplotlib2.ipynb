### Plotting a DataFrame

```python
import pandas as pd

data = {
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Visitors': [200, 230, 250, 180, 300],
    'Revenue': [1500, 1600, 1700, 1400, 2100]
}

df = pd.DataFrame(data)
```

```python
# line plot
df.plot(x='Day', y='Visitors', kind='line', marker='o')
plt.title("Visitors Per Day")
plt.xlabel("Day")
plt.ylabel("Visitors")
plt.grid(True)
plt.show()
```

```python
#bar chart
df.plot(x='Day', y='Revenue', kind='bar', color='orange')
plt.title("Revenue Per Day")
plt.ylabel("Revenue ($)")
plt.show()
```

```python
#multiple lines
df.plot(x='Day', y=['Visitors', 'Revenue'], kind='line', marker='o')
plt.title("Visitors & Revenue")
plt.ylabel("Count / $")
plt.legend()
plt.show()
```

```python
# scatter plot
df.plot(kind='scatter', x='Visitors', y='Revenue', color='red')
plt.title("Revenue vs Visitors")
plt.show()
```

```python
#pie chart
df.set_index('Day')['Visitors'].plot(kind='pie', autopct='%1.1f%%')
plt.title("Visitor Share")
plt.ylabel("")  # Hide y-label
plt.show()
```

```python
#Histogram
df['Revenue'].plot(kind='hist', bins=5, color='skyblue', edgecolor='black')
plt.title("Revenue Distribution")
plt.xlabel("Revenue")
plt.show()
```

### A figure is the entire canvas or window where one or more plots (also called Axes) can be drawn.
Detailed Anatomy (with Diagram Concept)
1. Figure (plt.figure or fig)
The top-level container.

Can hold multiple Axes.

You can set the overall size, dpi, and save the entire figure.

2. Axes (ax)
The actual plot area  where lines, bars, scatter points go.

One Figure can have many Axes.

Think of Axes as a single plot inside the figure.

3. Axis (xaxis, yaxis)
Part of the Axes, representing the horizontal and vertical scales.

4. Tick & TickLabel
Ticks are the marks along the axes.

Tick labels are the numbers or text at those ticks.

5. Labels
X-label and Y-label describe the axis units or data meaning.

6. Title
The heading for the entire plot (set with ax.set_title()).

7. Legend
Explains the meaning of colors/lines (ax.legend()).

8. Grid
Helps read values easily, controlled with ax.grid(True).

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(8, 4))  # 1. Create Figure and Axes

ax.plot(x, y, label='sin(x)')          # 2. Plot Data
ax.set_title("Sine Wave")              # 3. Set Title
ax.set_xlabel("X Axis (radians)")      # 4. Set X Label
ax.set_ylabel("Y Axis (amplitude)")    # 5. Set Y Label
ax.legend()                            # 6. Add Legend
ax.grid(True)                          # 7. Add Grid

plt.show()
```

```python
#Set Plot Title with Font Size

plt.plot([1, 2, 3], [10, 20, 30])
plt.title("My Awesome Plot", fontsize=16)  # Title and font size
plt.show()
```

```python
#Set Font Sizes for Everything using rcParams
plt.rcParams.update({
    "axes.titlesize": 20,
    "axes.labelsize": 14,
    "xtick.labelsize": 12,
    "ytick.labelsize": 12,
    "legend.fontsize": 12
})

plt.plot([1, 2, 3], [3, 6, 9], label='Line')
plt.title("Custom Global Font Sizes")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend()
plt.show()

```

```python
#Using Object-Oriented API (with ax)
fig, ax = plt.subplots()
ax.plot([1, 2, 3], [10, 20, 30])
ax.set_title("Plot Title", fontsize=16)
ax.set_xlabel("X Axis", fontsize=14)
ax.set_ylabel("Y Axis", fontsize=14)
plt.show()
```

```python
#Basic Line Styles

plt.plot([1, 2, 3, 4], [10, 20, 25, 30], linestyle='-')   # Solid line
plt.plot([1, 2, 3, 4], [15, 18, 22, 28], linestyle='--')  # Dashed line
plt.plot([1, 2, 3, 4], [5, 15, 10, 20], linestyle='-.')   # Dash-dot line
plt.plot([1, 2, 3, 4], [2, 5, 8, 12], linestyle=':')      # Dotted line
plt.title("Line Style Examples")
plt.legend(["Solid", "Dashed", "Dash-dot", "Dotted"])
plt.show()
```

### Line Styles Shortcuts in Format Strings
You can use format strings like "--r" to combine:

Line style

Marker style

Color

```python
plt.plot([1, 2, 3], [1, 4, 9], 'r--')  # Red dashed line
plt.plot([1, 2, 3], [2, 5, 10], 'g-.')  # Green dash-dot line
plt.plot([1, 2, 3], [3, 6, 11], 'b:')   # Blue dotted line
plt.title("Format String for Line Styles")
plt.legend(["Red Dashed", "Green Dash-dot", "Blue Dotted"])
plt.show()
```

```python
#Custom Line Width
#Add linewidth or lw:

plt.plot([1, 2, 3, 4], [3, 7, 12, 18], linestyle='--', linewidth=3)
plt.title("Custom Line Width")
plt.show()
```

```python
#Using Line Styles with Axes Object (OOP method)
fig, ax = plt.subplots()
ax.plot([0, 1, 2], [2, 3, 5], linestyle=':', linewidth=2)
ax.set_title("OOP Line Style Example")
plt.show()
```

```python
#Custom Line Styles (advanced)
line, = plt.plot([1, 2, 3, 4], [5, 10, 15, 20])
line.set_linestyle((0, (5, 10)))  # (offset, on_off_seq)  long dash
plt.title("Custom Dashed Line")
plt.show()
```

```python
#axis limits
#Using plt.xlim() and plt.ylim() (State-based approach)

x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

plt.plot(x, y)
plt.xlim(1, 4)   # Set x-axis limits
plt.ylim(0, 20)  # Set y-axis limits
plt.title("Axis Limits Example")
plt.show()

```

```python
# Using ax.set_xlim() and ax.set_ylim() (Object-Oriented approach)
fig, ax = plt.subplots()

ax.plot(x, y)
ax.set_xlim(1, 4)
ax.set_ylim(0, 20)
ax.set_title("OOP Axis Limits")
plt.show()
```

```python
#Setting Limits to Auto or None
plt.xlim(auto=True)  # or plt.xlim(None)
plt.ylim(auto=True)  # or plt.ylim(None)
```

```python
#Reverse the Axis
plt.xlim(5, 0)   # x-axis reversed
plt.ylim(30, 0)  # y-axis reversed
```

```python
#Get Current Axis Limits
x_limits = plt.xlim()
y_limits = plt.ylim()

print("X limits:", x_limits)
print("Y limits:", y_limits)
```

```python
#Set Figure Size Using plt.figure() (State-based)
plt.figure(figsize=(10, 5))  # width=10 inches, height=5 inches
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Figure Size Example")
plt.show()
```

```python
#Set Figure Size Using fig, ax = plt.subplots() (Object-Oriented)
fig, ax = plt.subplots(figsize=(8, 4))  # width=8 inches, height=4 inches
ax.plot([1, 2, 3], [4, 5, 6])
ax.set_title("OOP Figure Size")
plt.show()
```

```python
#Units Used for Figure Size
#figsize=(width, height) is in inches.
#Resolution can be controlled with DPI (dots per inch):

fig, ax = plt.subplots(figsize=(8, 6), dpi=100)
#this gives a figure that is 800x600 pixels.
```

```python
#customize axis ticks 
#Set Custom Tick Values

x = [0, 1, 2, 3, 4]
y = [10, 20, 25, 30, 40]

plt.plot(x, y)
plt.xticks([0, 2, 4])         # Show only ticks at x = 0, 2, 4
plt.yticks([10, 25, 40])      # Show only ticks at y = 10, 25, 40
plt.title("Custom Tick Values")
plt.show()
```

```python
#Set Custom Tick Labels
plt.plot(x, y)
plt.xticks([0, 1, 2, 3, 4], ['Zero', 'One', 'Two', 'Three', 'Four'])  # Custom labels
plt.yticks([10, 20, 30, 40], ['Low', 'Mid', 'High', 'Very High'])
plt.title("Custom Tick Labels")
plt.show()

```

```python
#Rotate Tick Labels
plt.plot(x, y)
plt.xticks(rotation=45)  # Rotate x-axis labels
plt.title("Rotated Tick Labels")
plt.show()
```

```python
#Set Custom Ticks Using Object-Oriented Interface
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xticks([1, 3])
ax.set_xticklabels(['One', 'Three'])
ax.set_yticks([20, 30])
ax.set_yticklabels(['Medium', 'High'])
ax.set_title("Ticks with OO Interface")
plt.show()
```

| Task               | Function                                   |
| ------------------ | ------------------------------------------ |
| Set tick positions | `plt.xticks([0, 1, 2])`                    |
| Set tick labels    | `plt.xticks([0,1], ['Zero', 'One'])`       |
| Rotate tick labels | `plt.xticks(rotation=45)`                  |
| Object-oriented    | `ax.set_xticks()` + `ax.set_xticklabels()` |


```python
#To draw vertical lines in a Matplotlib chart, you can use plt.axvline() (for procedural) or ax.axvline() (for object-oriented).
#Basic Vertical Line

plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.axvline(x=2, color='red', linestyle='--')  # Vertical line at x=2
plt.title("Vertical Line Example")
plt.show()
```

```python
#Multiple Vertical Lines
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
for xval in [1.5, 3.5]:
    plt.axvline(x=xval, color='blue', linestyle=':')
plt.title("Multiple Vertical Lines")
plt.show()
```

```python
#With Object-Oriented Interface
fig, ax = plt.subplots()
ax.plot(x, y)
ax.axvline(x=3, color='green', linewidth=2)
ax.set_title("OO Interface Vertical Line")
plt.show()
```

```python
 #With Labels (Annotated)
plt.plot(x, y)
plt.axvline(x=2, color='purple', linestyle='--')
plt.text(2.1, 20, 'Important Point', rotation=90, color='purple')
plt.title("Vertical Line with Label")
plt.show()
```

| Function                          | Description                     |
| --------------------------------- | ------------------------------- |
| `plt.axvline(x=...)`              | Add vertical line               |
| `color`, `linestyle`, `linewidth` | Customize appearance            |
| `plt.text()`                      | Add annotation next to the line |


```python
#plt.text(x, y, "text", ...)  add custom text at position
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.text(2, 22, "Peak Point", fontsize=12, color="red")
plt.title("Text Example")
plt.show()
```

```python
#plt.annotate()  advanced labeling with arrows
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.annotate("Important", 
             xy=(3, 25), 
             xytext=(2, 26), 
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=12)
plt.title("Annotation Example")
plt.show()
```

```python
#Object-Oriented Style (ax.text() / ax.annotate())
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [10, 20, 25, 30])
ax.text(1, 12, "Start", fontsize=10, color="blue")
ax.annotate("Max", xy=(3, 25), xytext=(2.5, 27),
            arrowprops=dict(facecolor='green', shrink=0.05))
plt.show()
```

| Parameter  | Description                       |
| ---------- | --------------------------------- |
| `x, y`     | Position of text                  |
| `s`        | The text string                   |
| `fontsize` | Size of text                      |
| `color`    | Color of text                     |
| `ha`, `va` | Horizontal/Vertical alignment     |
| `rotation` | Rotate the text (e.g. 90 degrees) |


```python
#annotate() lets you point to a location on the chart and place a text label nearby. 
#Its useful for highlighting important points like peaks, lows, or outliers.
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)

plt.annotate('Highest Point', 
             xy=(3, 25),             # Point to annotate
             xytext=(2.5, 27),       # Position of text
             arrowprops=dict(arrowstyle='->', color='red'),
             fontsize=12, color='blue')

plt.title("Annotation Example")
plt.grid(True)
plt.show()
```

```python
#Change arrow style and text box
plt.plot(x, y)

plt.annotate('Note Here',
             xy=(2, 20),
             xytext=(1.5, 24),
             arrowprops=dict(arrowstyle='fancy', 
                             facecolor='green', 
                             edgecolor='black'),
             bbox=dict(boxstyle="round,pad=0.3", 
                       edgecolor='black', facecolor='lightyellow'),
             fontsize=10)

plt.title("Fancy Annotation")
plt.show()
```

| Style     | Description            |            |                             |
| --------- | ---------------------- | ---------- | --------------------------- |
| `'->'`    | Standard arrow         |            |                             |
| `'fancy'` | Curved, stylized arrow |            |                             |
| `'wedge'` | Wedge-shaped arrow     |            |                             |
| \`'       | -                      | ', '-\['\` | Vertical lines and brackets |


```python
#Remove All Borders (Spines)
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

fig, ax = plt.subplots()
ax.plot(x, y)

# Remove all 4 spines
for spine in ['top', 'right', 'left', 'bottom']:
    ax.spines[spine].set_visible(False)

plt.title("Plot without Borders")
plt.grid(True)
plt.show()
```

```python
#Remove only top and right borders
fig, ax = plt.subplots()
ax.plot(x, y)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.title("Simplified Border Style")
plt.show()
```
